# Rota Bilgileri

```python
Rota1 = [["A", 100.], ["B", 75], ["C", 100], ["D", 75], ["E", 75.01] , ["F", 25] , ["O", 25] , ["F", 75], ["E", 75], ["D", 100] , ["C", 75], ["B", 100],["A", -1]]
Rota2 = [["G", 77.], ["H", 82.], ["I", 50.], ["F", 97.], ["J", 100.] , ["K", 112.], ["L", 100.], ["G", -1.]]
Rota3 = [["N", 100], ["K", 100], ["P", 75], ["R", 50], ["D", 50] , ["S", 75], ["P", 100],["K", 100.01], ["N", -1]]

rotalar = [Rota1, Rota2, Rota3]

def aradaki_mesafeyi_al(kalkis, varis, rota_indeks):
    secilen_rota = rotalar[rota_indeks]
    for i in range(len(secilen_rota)):
        if secilen_rota[i][0] == kalkis and secilen_rota[i+1][0] == varis:
            return secilen_rota[i][1]
    
```

# Trenler

```python
class Tren:
    """    tren_tipi
    #  0-> yuk, 1-> anahat, 2-> hizli
    
    hedef_durak
    kalkis_durak
    #sonrakine_kalan_mesafe
    # maksimum hızla gidildigini varsaydigimiz icin gereksiz
    sonraki_duraga_kalan_sure
    maks_hiz
    bakima_kalan_km
    durakta_kalan_bekleme_suresi
    toplanan_para
    tren_bakimda_mi
    """
    
    
    
    def __init__(self, tren_tipi, tren_ismi):
        
        global duraklar 
        global maks_hizlar
        global genel_bakim_km
        global durakta_bekleme_suresi
        global rota_indeksleri
        
        self.tren_ismi = tren_ismi
        self.tren_tipi = tren_tipi
        self.sefer_sayisi = 0
        self.hedef_durak = rotalar[rota_indeksleri[tren_tipi]][1]
        self.kalkis_durak = rotalar[rota_indeksleri[tren_tipi]][0] #dictionary 
        self.sonraki_duraga_kalan_sure = rotalar[rota_indeksleri[tren_tipi]][0][1] / maks_hizlar[tren_tipi]
        self.maks_hiz = maks_hizlar[tren_tipi]
        self.bakima_kalan_km = genel_bakim_km[tren_tipi]
        self.durakta_bekleme_suresi = durakta_bekleme_suresi[tren_tipi]
        self.toplanan_para = 0
        self.tren_bakimda_mi = False
        self.bakim_bitmesine_sure = 0
        self.tablo = Tablo(tren_tipi, tren_ismi, 0)
        durak_bekleme_suresi_ekle(5,  rotalar[rota_indeksleri[tren_tipi]][1][0])
        
        
def cakisma_kontrol(isim , varisa_dakika,):
    donulecek_trenler = []
    for i in range(len(trenler)):
        if trenler[i].hedef_durak == isim and trenler[i].sonraki_duraga_kalan_sure == varisa_dakika and not trenler[i].tren_bakimda_mi:
            donulecek_trenler.append(i)
            
    return donulecek_trenler
            
    
```

# Duraklar

```python
class Durak:
    
    def __init__(self, isim):
        self.isim = isim
        self.mesgul_suresi = 0
        self.su_an_kim_var = ""
        
        
def duraklari_init():
    duraklar = []
    isimler = ["A", "B", "C", "D", "E", "F", "O", "G", "H", "I", "F", "J", "K", "L", "N", "P", "R", "S" ]
    
    for isim in isimler:
        temp = Durak(isim)
        duraklar.append(temp)
        
    return duraklar



duraklar = duraklari_init()


def get_durak_mesgul(isim):
    for Durak in duraklar:
        if Durak.isim == isim:
            return Durak.mesgul_suresi
        


def durak_bekleme_suresi_ekle(sure, durak_ismi):
    for Durak in duraklar:
        if Durak.isim == durak_ismi:
            Durak.mesgul_suresi = sure
        return
    
def set_durak_kim_var(tren_ismi, durak_ismi):
    for Durak in duraklar:
        if Durak.isim == durak_ismi:
            Durak.su_an_kim_var = tren_ismi
        return
    
```

```python

## duragina en yakin tren
def  tren_secimi():
    
    en_yakin_sure = 9999999999
    en_yakin_index = -1
    
    for i in range(len(trenler)):
        
        if trenler[i].sonraki_duraga_kalan_sure + trenler[i].durakta_bekleme_suresi +  trenler[i].bakim_bitmesine_sure < en_yakin_sure and not trenler[i].tren_bakimda_mi:
            en_yakin_index = i
            en_yakin_sure = trenler[i].sonraki_duraga_kalan_sure + trenler[i].durakta_bekleme_suresi +  trenler[i].bakim_bitmesine_sure
            
    return en_yakin_index
```
